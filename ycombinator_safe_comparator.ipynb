```python
from docx import Document
import difflib
from google.colab import files

# Funci√≥n para leer un archivo .docx como l√≠neas de texto
def leer_docx_como_lineas(nombre_archivo):
    doc = Document(nombre_archivo)
    return [p.text.strip() for p in doc.paragraphs if p.text.strip()]
```

```python
# Sub√≠ el archivo base original de YCombinator una vez
print("üì§ Sub√≠ el TEMPLATE SAFE oficial (solo una vez)")
uploaded_template = files.upload()
template_filename = next(iter(uploaded_template))
template_lineas = leer_docx_como_lineas(template_filename)

print(f"‚úÖ Template cargado: {template_filename}")
```

```python
def comparar_con_template(template_lineas):
    print("üì§ Sub√≠ un contrato SAFE revisado para comparar con el template")
    uploaded_revisado = files.upload()
    revisado_filename = next(iter(uploaded_revisado))
    revisado_lineas = leer_docx_como_lineas(revisado_filename)

    print(f"\nüìä Comparando: {revisado_filename} con el template base...\n")

    d = difflib.Differ()
    diff = list(d.compare(template_lineas, revisado_lineas))

    for linea in diff:
        if linea.startswith("- "):
            print(f"‚ùå ELIMINADA o CAMBIADA del template: {linea[2:]}")
        elif linea.startswith("+ "):
            print(f"‚ûï AGREGADA en el revisado: {linea[2:]}")
```

```python
comparar_con_template(template_lineas)
```
